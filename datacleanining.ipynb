```python
import pandas as pd


df = pd.read_csv("data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure size
plt.figure(figsize=(15, 10))

# Subplot 1: Pulse
plt.subplot(2, 2, 1)
sns.histplot(df['Pulse'], bins=30, kde=True)
plt.title('Pulse Distribution')

# Show the plot
plt.tight_layout()
plt.show()
```
